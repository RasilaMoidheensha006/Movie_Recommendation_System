```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
#creating the ratings dataframe with average rating and number of ratings
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
#creating the ratings dataframe with count of rating and number of ratings
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

setting the number of rating column

```python
ratings['number of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#histogram
plt.figure(figsize=(10,4))
ratings['number of ratings'].hist(bins=70)
```

```python

plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
#joinplot
sns.jointplot(x='rating',y='number of ratings',data=ratings,alpha=0.5)
```

Recommending Similiar Movies

```python
moviemat=df.pivot_table(index='userId',columns='title',values='rating')
moviemat.head(10345)
```

```python
moviemat.isnull().sum()
```

```python
#most rated movies
ratings.sort_values('number of ratings',ascending=False).head(10)
```

```python
#grabing the user for rating for those movies
Pulp_fiction_rating=moviemat['Pulp Fiction (1994)']
Jurassic_park_rating=moviemat['Jurassic Park (1993)']
Pulp_fiction_rating.head()
```

```python
#to get the correlation between two pandas series
similar_to_pulp=moviemat.corrwith(Pulp_fiction_rating)
similar_to_jur=moviemat.corrwith(Jurassic_park_rating)
```

Cleaning the NaN values by removing them and using a Datframe instead of a series


```python
corr_pulp=pd.DataFrame(similar_to_pulp,columns=['Correlations'])
corr_pulp.dropna(inplace=True)
corr_pulp.head()
```

```python
#joining value
corr_pulp=corr_pulp.join(ratings['number of ratings'],lsuffix='_left')
corr_pulp.head()
```

```python
corr_pulp.drop(columns=['number of ratings_left'],inplace=True)
```

```python
corr_pulp
```

```python
#sorting the correlation values
corr_pulp.sort_values('Correlations',ascending=False).head(10)
```

```python
#now sort  values and notice how the title make a lot more sense
corr_pulp[corr_pulp['number of ratings']>100].sort_values('Correlations',ascending=False).head()
```

```python
#doing the same for Jurrasic park
corr_jur=pd.DataFrame(similar_to_jur,columns=['Correlations'])
corr_jur.dropna(inplace=True)
corr_jur.head()
```

```python
#sorting the correlation values
corr_jur.sort_values('Correlations',ascending=False).head(10)
```

```python
corr_jur=corr_pulp.join(ratings['number of ratings'],lsuffix='_left')
corr_jur.head()
```

```python
corr_jur.drop(columns=['number of ratings_left'],inplace=True)
```

```python
corr_jur
```

```python
#now sort  values and notice how the title make a lot more sense
corr_jur[corr_jur['number of ratings']>100].sort_values('Correlations',ascending=False).head()
```
